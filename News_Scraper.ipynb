```python
from bs4 import BeautifulSoup
from newspaper import Article
import nltk
import pandas as pd
import requests
nltk.download('punkt')

url = "https://news.google.com/topics/CAAqJggKIiBDQkFTRWdvSUwyMHZNRGx6TVdZU0FtVnVHZ0pKVGlnQVAB?hl=en-IN&gl=IN&ceid=IN%3Aen"

req = requests.get(url)
coverpage = req.content

soup1 = BeautifulSoup(coverpage, 'html5lib')
coverpage_news = soup1.find_all('div', attrs = {'class':'DBQmFf NclIid BL5WZb Oc0wGc xP6mwf j7vNaf'})
print("Number of news articles found = ",len(coverpage_news))
length = len(coverpage_news)

main_news_title = []
main_news_link = []
main_news_date = []
main_news_summary = []

sub_news_title = []
sub_news_link = []
sub_news_date = []
sub_news_summary = []

# displaying data for only the top 10 news articles..

for index, ele in enumerate(coverpage_news[:10]):
  print("\nARTICLE NO : ",index+1)
  
  # MAIN NEWS - DATA EXTRACTION
  print("\nMAIN NEWS :- ")
  main_news = ele.find('a', class_='DY5T1d')
  href = "http://news.google.com" + main_news['href'][1:]
  print("Found an URL :", href)
  main_news_link.append(href)

  main_article = Article(href, language="en") # en for English 
  main_article.download()
  main_article.parse()
  main_article.nlp()
  
  print("Title : \n\t",main_article.title)
  main_news_title.append(main_article.title)

  #print("Article Text : \n", main_article.text)
  print("Summary : \n\t", main_article.summary)
  main_news_summary.append(main_article.summary)
  print(main_article.keywords)  

  print("Publish Date : ", main_article.publish_date)
  main_news_date.append(main_article.publish_date)
  print()
  
  # SUB NEWS - DATA EXTRACTION
  print("SUB NEWS : ")
  news = ele.find('h4', class_='ipQwMb ekueJc RD0gLb')
  sub_news = news.find('a', class_='DY5T1d RZIKme')

  href = "http://news.google.com" + sub_news['href'][1:]
  print("Found an URL :", href)
  sub_news_link.append(href)

  sub_article = Article(href, language="en") # en for English 
  sub_article.download()
  sub_article.parse()
  sub_article.nlp()
  
  print("Title : \n\t",sub_article.title)
  sub_news_title.append(sub_article.title)

  #print("Article Text : \n", sub_article.text)
  print("Summary : \n\t", sub_article.summary)
  sub_news_summary.append(sub_article.summary)
  
  print("Publish Date : ", sub_article.publish_date)
  sub_news_date.append(sub_article.publish_date)
  print()
```

```python
main_news_db = pd.DataFrame({
    'Main News Title' : main_news_title,
    'Summarised News Text' : main_news_summary,
    'URL' : main_news_link,
    'Publish Datetime' : main_news_date 
})
print(main_news_db.head())

sub_news_db = pd.DataFrame({
    'Sub News Title' : sub_news_title,
    'Summarised News Text' : sub_news_summary,
    'URL' : sub_news_link,
    'Publish Datetime' : sub_news_date 
})
print(sub_news_db.head())
```

```python
def search_word():  
  word = input("Enter a word to be searched : ")
  print(word)

  for i in range(10):
    main_news_obj = main_news_db.iloc[[i]]
    main_url = main_news_obj.iloc[0]['URL']
    
    article = Article(main_url, language="en") # en for English 
    article.download()
    article.parse()
    article.nlp()

    text = article.text
    text = text.split('\n')
    final_text = []             # final list of sentences formed from the text of the article
    for ind in range(len(text)):
      final= text[ind].split(". ")
      final_text.append(final)
    # print(final_text)

    flag = 0
    line_list = []            # list of line containing the word
    for line in text:
      if word in line:
        flag = 1
        line_list.append(line)

    if flag:
      print("\nCHECKING ARTICLE NO : ",i+1, "...")
      print("\n\t IN MAIN NEWS MATCHING ARTICLES ARE :- \n")
      for ele in line_list:
        print(" -> ", ele)
        
  # if word is found in main article only then is it searched in the sub article..
  
      sub_news_obj = sub_news_db.iloc[[i]]
      sub_url = sub_news_obj.iloc[0]['URL']
    
      article = Article(sub_url, language="en") 
      article.download()
      article.parse()
      article.nlp()
      text = article.text
      text = text.split('\n')
      final_text_sub = []             # final list of sentences formed from the text of the article
      for ind in range(len(text)):
        final= text[ind].split(". ")
        final_text_sub.append(final)
      # print(final_text_sub)

      flag_sub = 0
      line_list_sub = []
      for line in text:
        if word in line:
          flag_sub = 1 
          line_list_sub.append(line)
      if flag_sub:
        print("\n\tIN SUB NEWS MATCHING LINES ARE :- \n")
        for ele in line_list_sub:
          print(" -> ", ele)
    else:
      print("\nCHECKED ARTICLE NO :",i+1,"... Match not found !! ...")
      
search_word()
```
